```python
import duckdb

# Connect to (or create) your database
con = duckdb.connect("instacart.duckdb")

# Load products CSV into a table
con.execute("""
    CREATE OR REPLACE TABLE products AS
    SELECT * FROM read_csv_auto('products.csv');
""")

# Run query and store result as DataFrame
result_df = con.execute("""
    SELECT *
    FROM products
    LIMIT 100;
""").fetchdf()

# Show the top rows
result_df.head()
print("~~~~~~~~~~")
result_df.info()
print("~~~~~~~~~~")
result_df.describe()

```
